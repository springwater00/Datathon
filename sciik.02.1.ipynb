```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
mat = pd.read_csv("student-mat.csv")
por = pd.read_csv("student-por.csv")

mat["subject"] = "mat"
por["subject"] = "por"

data = pd.concat([mat, por])
data["avg_G"] = (data["G1"] + data["G2"]) / 2
data
```

```python
plt.rc('font', family='NanumGothic')
```

## 사교육과 성적 알콜소비와의 관계비교

```python
temp = data["schoolsup"].value_counts()
plt.bar(x=np.arange(2), height=temp.values)
plt.xticks(np.arange(2), temp.index)
plt.title("전체 사교육 수강 비교")
plt.show()

temp = data.groupby("subject")["schoolsup"].value_counts()

plt.subplot(121)
plt.bar(x=np.arange(2), height=temp["mat"].values)
plt.xticks(np.arange(2), temp["mat"].index)
plt.title("수학과목의 사교육 수강 비교")

plt.subplot(122)
plt.bar(x=np.arange(2), height=temp["por"].values)
plt.xticks(np.arange(2), temp["por"].index)
plt.title("포루투갈어과목의 사교육 수강 비교")

plt.tight_layout()
plt.show()

fig = plt.figure(figsize=(12,5))
plt.subplot(131)
sns.scatterplot(data=data, x="Dalc", y="G1", hue="schoolsup")
plt.title("사교육 수강 여부에 따른 평일 알콜섭취와 G1성적비교")

plt.subplot(132)
sns.scatterplot(data=data, x="Dalc", y="G2", hue="schoolsup")
plt.title("사교육 수강 여부에 따른 평일 알콜섭취와 G2성적비교")

plt.subplot(133)
sns.scatterplot(data=data, x="Dalc", y="G3", hue="schoolsup")
plt.title("사교육 수강 여부에 따른 평일 알콜섭취와 G3성적비교")

plt.tight_layout()
plt.show()

print()
print()
print()

fig = plt.figure(figsize=(12,5))
plt.subplot(131)
sns.scatterplot(data=data, x="Walc", y="G1", hue="schoolsup")
plt.title("사교육 수강 여부에 따른 주말 알콜섭취와 G1성적비교")

plt.subplot(132)
sns.scatterplot(data=data, x="Walc", y="G2", hue="schoolsup")
plt.title("사교육 수강 여부에 따른 주말 알콜섭취와 G2성적비교")

plt.subplot(133)
sns.scatterplot(data=data, x="Walc", y="G3", hue="schoolsup")
plt.title("사교육 수강 여부에 따른 주말 알콜섭취와 G3성적비교")

plt.tight_layout()
plt.show()
```
