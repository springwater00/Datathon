```python
import pandas as pd             # data analysis
import numpy as np              # linear algebra + array handling
import sklearn                  # machine learning
import matplotlib.pyplot as plt # visualization
import seaborn as sns  
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
sns.set_palette('deep')
plt.rc('font', family='NanumGothic')
```

```python
mat_data = pd.read_csv('student-mat.csv')
por_data = pd.read_csv('student-por.csv')
df = pd.concat([mat_data,por_data])
mat_data['subject'] = 'mat'
por_data['subject'] = 'por'
df = pd.concat([mat_data,por_data])
```

# 1. Failures 

```python
df["failures"].value_counts()
plt.pie(df["failures"].value_counts().values, labels=df["failures"].value_counts().index, autopct="%d%%")
plt.title('F   ')
plt.show()

ax = sns.countplot(x = 'failures',data = df)
containers = ax.containers[0]
ax.bar_label(containers, labels=[f'{x:,.0f}' for x in containers.datavalues], label_type='center')
plt.title('F   ')
plt.show()

ax = sns.countplot(x = 'failures',data = df,hue = 'subject')
containers = ax.containers[0]
ax.bar_label(containers, labels=[f'{x:,.0f}' for x in containers.datavalues], label_type='center')
containers1 = ax.containers[1]
ax.bar_label(containers1, labels=[f'{x:,.0f}' for x in containers1.datavalues], label_type='center')
plt.title('F    ')
plt.show()


ax = sns.violinplot(data=df, x="failures", y="G1")
plt.title('F  G1 ')
plt.show()

ax = sns.violinplot(data=df, x="failures", y="G2")
plt.title('F  G2 ')
plt.show()

```
